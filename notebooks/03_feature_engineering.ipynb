```python
# Dropping BUI column to prevent multicollinearity
# Dropping year columns as it only contains year 2012 which has no significant effect on our model
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Load the cleaned dataset
df = pd.read_csv("../data/processed/cleaned_algerian_fire_data.csv")

# 2. Drop unnecessary or highly multicollinear columns
# 'year' has zero variance (all 2012). 'BUI' is 98% correlated with 'DMC'.
df = df.drop(columns=['year', 'BUI'])

# 3. Separate features (X) and target label (y)
X = df.drop(columns=['Classes'])
y = df['Classes']

# 4. Split the data into Training and Testing sets (80% train, 20% test)
# stratify=y ensures our fire/no-fire ratio stays identical in both splits
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```

```python
# Initialize the scaler
scaler = StandardScaler()

# CRITICAL: Fit and transform ONLY on the training data to avoid data leakage
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using the weights learned from the training data
X_test_scaled = scaler.transform(X_test)

# Convert back to clean DataFrames for easier viewing/tracking
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

X_train_scaled.head()
```

```python
# Save the engineered training and test splits to CSVs
X_train_scaled.to_csv("../data/processed/X_train_scaled.csv", index=False)
X_test_scaled.to_csv("../data/processed/X_test_scaled.csv", index=False)
y_train.to_csv("../data/processed/y_train.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)

print("All training and test partitions successfully saved to data/processed!")
```
